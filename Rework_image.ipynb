```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from ipywidgets import interact, interactive, fixed, interact_manual, widgets
%matplotlib inline
from dep.fonctions import crop_image
```

# Pour cropper les images

```python
imgs =glob.glob('assets/img/*.png')
for i, im in enumerate(imgs):
    if ('3D' in im) and 'cropped' not in im:
        print(i, im)
```

```python
numero_image = 4

top_horiz = widgets.IntSlider(
    value=500,
    min=0,
    max=1080,
    step=10,
    description='Top:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
bottom_horiz = widgets.IntSlider(
    value=1440,
    min=0,
    max=1440,
    step=10,
    description='Bottom:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
left_vert = widgets.IntSlider(
    value=500,
    min=0,
    max=1920,
    step=10,
    description='Left:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
right_vert = widgets.IntSlider(
    value=1650,
    min=0,
    max=1920,
    step=10,
    description='Right:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

# Create a function to show the image with horizontal red lines
def show_image(top, bottom, left, right):
    img = mpimg.imread(imgs[numero_image])
    plt.imshow(img)
    plt.axhline(y=top, color='r', linestyle='-', linewidth=2)
    plt.axhline(y=bottom, color='r', linestyle='-', linewidth=2)
    plt.axvline(x=left_vert.value, color='r', linestyle='-', linewidth=2)
    plt.axvline(x=right_vert.value, color='r', linestyle='-', linewidth=2)
    plt.show()

interact(show_image, top=top_horiz, bottom=bottom_horiz, left=left_vert, right=right_vert)
```

```python
crop_image(img_path=imgs[numero_image], top=top_horiz.value, bottom=bottom_horiz.value, left=left_vert.value, right=right_vert.value, show=True)
```
